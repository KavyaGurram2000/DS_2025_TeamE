```python
import pandas as pd

# Load the dataset
file_path = "County_zhvi_bdrmcnt_2_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv"  # Change if needed
df = pd.read_csv(file_path)

# Display basic information about the dataset
df.info(), df.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum().sum()
print(f"Total missing values: {missing_values}")

# Summary statistics
df.describe()
```

```python
import matplotlib.pyplot as plt

# Select a few counties for visualization
selected_counties = ["Los Angeles County", "Cook County", "Harris County", "Maricopa County", "San Diego County"]
df_selected = df[df["RegionName"].isin(selected_counties)]

# Extract time-series data
df_time_series = df_selected.set_index("RegionName").iloc[:, 9:].T

# Plot trends for selected counties
plt.figure(figsize=(12, 6))
for county in selected_counties:
    plt.plot(df_time_series.index, df_time_series[county], label=county)

plt.xlabel("Year")
plt.ylabel("Property Value ($)")
plt.title("Property Value Trends Over Time (Selected Counties)")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

# Extract the latest available property values (Jan 2025)
latest_values = df["2025-01-31"].dropna()

# Plot histogram and density plot
plt.figure(figsize=(10, 5))
sns.histplot(latest_values, bins=50, kde=True)
plt.xlabel("Property Value ($)")
plt.ylabel("Frequency")
plt.title("Distribution of Property Values (January 2025)")
plt.show()
```

```python
# Compute the correlation matrix for the most recent values
correlation_matrix = df.iloc[:, 9:].corr()

# Plot heatmap of correlations
plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix, cmap="coolwarm", center=0, linewidths=0.5)
plt.title("Correlation Matrix of Property Values Across Counties")
plt.show()
```

```python
# Extract month-wise average property values
df_time_series["Month"] = df_time_series.index.str[5:7]  # Extract month from date
monthly_avg = df_time_series.groupby("Month").mean().mean(axis=1)

# Plot seasonal trends
plt.figure(figsize=(10, 5))
monthly_avg.plot(marker="o")
plt.xlabel("Month")
plt.ylabel("Average Property Value ($)")
plt.title("Seasonal Trends in Property Values")
plt.xticks(range(0, 12), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.grid()
plt.show()
```

```python
from fpdf import FPDF

# Create a PDF report
pdf = FPDF()
pdf.set_auto_page_break(auto=True, margin=15)
pdf.add_page()
pdf.set_font("Arial", style="", size=12)

# Title
pdf.set_font("Arial", style="B", size=16)
pdf.cell(200, 10, "Exploratory Data Analysis (EDA) Report", ln=True, align="C")
pdf.ln(10)

# Section: Overview
pdf.set_font("Arial", style="B", size=14)
pdf.cell(0, 10, "1. Dataset Overview", ln=True)
pdf.set_font("Arial", size=12)
pdf.multi_cell(0, 8,
    "- The dataset contains 2,531 rows and 310 columns.\n"
    "- It includes county-level property values from 2000 to 2025.\n"
    "- Some missing values (213,101) are present, especially in older records.\n"
)

# Section: Trends
pdf.set_font("Arial", style="B", size=14)
pdf.cell(0, 10, "2. Trends in Property Values", ln=True)
pdf.set_font("Arial", size=12)
pdf.multi_cell(0, 8,
    "- Property values have generally increased over time.\n"
    "- Los Angeles and San Diego Counties have the highest values, while Harris County has the lowest.\n"
    "- The dataset reflects economic cycles, with periodic dips.\n"
)

# Section: Distribution
pdf.set_font("Arial", style="B", size=14)
pdf.cell(0, 10, "3. Distribution of Property Values (Jan 2025)", ln=True)
pdf.set_font("Arial", size=12)
pdf.multi_cell(0, 8,
    "- The distribution is right-skewed, meaning most values are low, with a few high-value outliers.\n"
    "- Most properties are valued under $500,000.\n"
    "- Significant regional disparities exist in property values.\n"
)

# Section: Correlation
pdf.set_font("Arial", style="B", size=14)
pdf.cell(0, 10, "4. Correlation Analysis", ln=True)
pdf.set_font("Arial", size=12)
pdf.multi_cell(0, 8,
    "- There is a strong correlation between property values in different counties.\n"
    "- This suggests that economic and market factors impact regions similarly.\n"
)

# Section: Seasonality
pdf.set_font("Arial", style="B", size=14)
pdf.cell(0, 10, "5. Seasonal Trends", ln=True)
pdf.set_font("Arial", size=12)
pdf.multi_cell(0, 8,
    "- Prices peak in summer (June-July) and dip in winter (October-January).\n"
    "- This trend reflects increased home-buying activity in warmer months.\n"
)

# Save the report
pdf_file_path = "EDA_Report.pdf"
pdf.output(pdf_file_path)
print(f"PDF Report Saved: {pdf_file_path}")
```
